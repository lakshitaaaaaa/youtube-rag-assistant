Building a Chain

since each method/function is being called separately
-> build a chain.

this chaiin will be build by combining two chains :
  1. prompt template requires two inputs - context + question

  question : given by user, thus can be sent directly.

  context : retriver + question -> (from. vector store) gives context

  2 paralllel chains - context and question

  2. prompt + llm -> response
    chain : prompt + llm + parser

    linear chain

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrieved_docs):
  context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
  return context_text
```

```python
parallel_chain = RunnableParallel({
    'context' : retriever | RunnableLambda(format_docs),
    'question' : RunnablePassthrough()
})

# first chain -> runnableParallel mai we need to passs a dictionry
# when a question arrives we pass the question to the retriver, and via semanic searching in the vector database
# a context is retrived in the form of list of string, which is then passed to format_docs function
# which returns the final formatted context.

# second is question, the question which was the input is sent as the output, therefore RunnablePassThrough()

# thus the first parallel chain has been generated
```

```python
parser = StrOutputParser()
```

```python
main_chain = parallel_chain | prompt | llm | parser

# in this, 3 things will be presnt : prompt , llm, parser
# this chain will then be connected to parallel_chain
```

```python
main_chain.invoke('Can you summarize the video?')
```
